# Introduction

in this project i will create an imdb scraper to get information about the top titles sorted by the number of votes that has been released between **2000-01-01** and **2018-12-31**, I'm going to extract data such as votes, meta score and year, I will extract the data needed using a library called **BeautifulSoup** and to get information about a website we use get in the **requests** library.

# Importing the required libraries

````python
import pandas as pd
import numpy as np
from requests import get # Retrieving information in a website
from bs4 import BeautifulSoup # Extracting data
from time import time # calculate the time
from time import sleep # pause the loop
from random import randint # Randomize sleep time
from IPython.core.display import clear_output # Stack the output
from warnings import warn # Warn when something goes wrong
````

# Getting the url

Retrieving informations about the url and checking if it's working.

````python
# Store the url we will get data from
url = 'https://www.imdb.com/search/title?release_date=2018-01-01,2018-12-31&sort=num_votes,desc&start=01&ref_=adv_nxt'
# Request for the content of the url
response = get(url)
# Print out to check if it work
print(response.text[:1000])
````

````python
# Parsing response to HTML
html = BeautifulSoup(response.text, 'html.parser')
# Print out the html to understand the structure and retrive data.
print(html)
````

## Creating a test loop

#### Extract data from each container and put them in their list

````python
names = []
years = []
imdb_ratings = []
metascores = []
votes = []

# Extracting data from titles
for container in title_containers:

    # Getting the name
    name = container.h3.a.text
    names.append(name)

    # Getting the year
    year = container.h3.find('span', class_ = 'lister-item-year').text
    years.append(year)

    # Getting the rating
    imdb = float(container.strong.text)
    imdb_ratings.append(imdb)

    # Getting the Metascore if the value exist
    if container.find('div', class_ = 'ratings-metascore') is not None:
        m_score = container.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))
    # Else assign in the NaN
    else:
        m_null = np.NaN
        metascores.append(m_null)

    # Getting the number of votes
    vote = container.find('span', attrs = {'name':'nv'})['data-value']
    votes.append(int(vote))
````

#### Test to check if everything works

````python
test_df = pd.DataFrame({'movie': names,
                       'year': years,
                       'imdb': imdb_ratings,
                       'metascore': metascores,
                       'votes': votes})
print(test_df.info())
test_df
````

## Creating the scaper loop for movies between 2000 and 2018.

#### scrape 6 pages in each year (300 titles)

````python
pages = [str(i) for i in range(1,252,50)]
pages
````

#### Scrape titles released between 2000-01-01 and 2018-12-31

````python
years_url = [str(i) for i in range(2000,2019)]
years_url
````

#### Pause the loop to sleep for 8 to 15 seconds randomly and observe the requests speed to not crash the server

````python
start_time = time()
requests = 0

for _ in range(5):
    # Pause the loop for 8~15 seconds
    sleep(randint(8,15))
    # Observing request/speed
    requests += 1
    current_time = time()
    elapsed_time = current_time - start_time
    # Print out request/speed
    print('Request: {}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    # Stack the output
    clear_output(wait = True)
````

#### Putting everything together

````python
names = []
years = []
imdb_ratings = []
metascores = []
votes = []

start_time = time()
requests = 0

for year_url in years_url:

    # For every page 
    for page in pages:

        # Make a get request
        response = get('https://www.imdb.com/search/title?release_date=' + year_url + '-01-01,' + year_url +
                       '-12-31&sort=num_votes,desc&start=' + page + '&ref_=adv_nxt')

        # Pause the loop for 8~15 seconds
        sleep(randint(8,15))

        # observe the requests
        requests += 1
        elapsed_time = time() - start_time
        print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
        # Stack the output
        clear_output(wait = True)

        # Put a warning if != 200 status codes
        if response.status_code != 200:
            warn('Request: {}; Status code: {}'.format(requests, response.status_code))

        # Break the loop if the number of requests is over 210
        if requests > 210:
            warn('Number of requests was greater than expected.')  
            break 

        # Parse the url with BeautifulSoup
        page_html = BeautifulSoup(response.text, 'html.parser')

        # Creating movie container for all 50 titles
        title_containers = page_html.find_all('div', class_ = 'lister-item mode-advanced')

        # For every title
        for container in title_containers:
            
            # Getting the name
            name = container.h3.a.text
            names.append(name)

            # Getting the year
            year = container.h3.find('span', class_ = 'lister-item-year').text
            years.append(year)

            # Getting the rating
            imdb = float(container.strong.text)
            imdb_ratings.append(imdb)

            # Getting the Metascore if the value exist
            if container.find('div', class_ = 'ratings-metascore') is not None:
                m_score = container.find('span', class_ = 'metascore').text
                metascores.append(int(m_score))
            # Else assign in the NaN
            else:
                m_null = np.NaN
                metascores.append(m_null)

            # Getting the number of votes
            vote = container.find('span', attrs = {'name':'nv'})['data-value']
            votes.append(int(vote))

end_time = time()
# Print out how many seconds it took to run the loop
print(end_time - start_time)
````
